```python
flower = {'Setosa' :0,'Versicolor':1,'Virginica':2}
iris['variety'] = iris['variety'].map(flower)
```

```python
iris.head()
```

```python
iris
```

**Preparing inputs and outputs**

```python
x = iris[['sepal.length','sepal.width',	'petal.length',	'petal.width']].values
y = iris[["variety"]].values
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()#Initialization
```

```python
model.fit(x,y)#Training the model
```

**Accuracy**

```python
model.score(x,y)
```

make predictions

```python
expected = y
predicted = model.predict(x)
predicted
```

To summarize the fit of the model

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```

**Separating the input and output columns**

```python
data = iris.values
x = data[:,0:4]
y = data[:,4]
```

```python
print(x)
```

Split the data to train and test the dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
print(x_train)
```

```python
print(y)
```

```python
print(y_train)
```

```python
print(y_test)
```

Support vector machine algorithm

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC()
```

```python
model_svc.fit(x_train,y_train)
```

```python
prediction_svc = model_svc.predict(x_test)
print(prediction_svc)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,prediction_svc))
for i in range(len(prediction_svc)):
  print(y_test[i],prediction_svc[i])
```

```python
model_svc.score(x_train,y_train)
```
